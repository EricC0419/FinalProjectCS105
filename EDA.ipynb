```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("drugLibTrain_raw.csv")

Douma = df['urlDrugName'].value_counts().head(12)

plt.figure(figsize=(10,5))
Douma.plot(kind='bar', color = 'red')

plt.title('Most reviewed drugs')
plt.xlabel('Drug name')
plt.ylabel('Number of reviews')
plt.xticks(rotation=45, ha='right')
plt.grid(axis = 'y', linestyle = '-', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("drugLibTrain_raw.csv")

conditions = df['condition'].value_counts().head(10)

plt.figure(figsize=(10,6))
conditions.plot(kind='bar')
plt.title('Most common conditions during reviews')
plt.xlabel('condition')
plt.ylabel('number of reviews')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
